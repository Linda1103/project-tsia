```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from scipy.optimize import curve_fit

plt.rcParams['font.sans-serif'] = ["SimHei"]
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['axes.unicode_minus'] = False
```

```python
zq = pd.read_excel("各洲确诊.xlsx")
zq.head()
```

```python
t = list(range(1,1 + len(zq),1))
t = np.array(t).astype(int)
```

```python
y = np.array(list(zq["欧洲"])).astype(int)
z = np.array(list(zq["北美洲"])).astype(int)
k = np.array(list(zq["南美洲"])).astype(int)
d = np.array(list(zq["大洋洲"])).astype(int)
f = np.array(list(zq["非洲"])).astype(int)
m = np.array(list(zq["亚洲"])).astype(int)
```

```python
print("时间t :\n",t)
#print("确诊人数 :\n",y,z,k,d,f)
```

```python
f1 = np.polyfit(t,y,3)
f2 = np.polyfit(t,z,3)
f3 = np.polyfit(t,k,3)
f4 = np.polyfit(t,d,3)
f5 = np.polyfit(t,f,3)
f6 = np.polyfit(t,m,3)
```

```python
print("a,b,c,d = :\n",f1)
print("a,b,c,d = :\n",f2)
print("a,b,c,d = :\n",f3)
print("a,b,c,d = :\n",f4)
print("a,b,c,d = :\n",f5)
print("a,b,c,d = :\n",f6)
```

```python
# 返回拟合函数
p1 = np.poly1d(f1)
p2 = np.poly1d(f2)
p3 = np.poly1d(f3)
p4 = np.poly1d(f4)
p5 = np.poly1d(f5)
p5 = np.poly1d(f6)
```

```python
print("多项式曲线拟合函数 :\n",p1,p2,p3,p4,p5)
```

```python
yvals = np.polyval(f1,t)
zvals = np.polyval(f2,t)
kvals = np.polyval(f3,t)
dvals = np.polyval(f4,t)
avals = np.polyval(f5,t)
mvals = np.polyval(f6,t)
```

```python
plot1 = plt.plot(t,z,"ob")
plot2 = plt.plot(t,zvals,"r",label="北美洲")
plot1 = plt.plot(t,y,"ob")
plot2 = plt.plot(t,yvals,"g",label="欧洲")
#plot1 = plt.plot(t,k,"ob")
plot2 = plt.plot(t,kvals,"b",label="南美洲")
#plot1 = plt.plot(t,d,"ob")
plot2 = plt.plot(t,dvals,"y",label="大洋洲")
#plot1 = plt.plot(t,a,"ob")
plot2 = plt.plot(t,avals,"k",label="非洲")
plot1 = plt.plot(t,m,"ob")
plot2 = plt.plot(t,mvals,"m",label="亚洲")
plt.xlabel("t (t0=27)")
plt.ylabel("各洲的确诊人数的多项式拟合")
plt.grid(False)
plt.legend()
plt.show()
```
